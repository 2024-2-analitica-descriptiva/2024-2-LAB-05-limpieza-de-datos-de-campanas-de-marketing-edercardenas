```python
import pandas as pd
import zipfile
import os
import numpy as np
```

```python
file_names = os.listdir('../files/input/')

for i, file in enumerate(file_names):
    with zipfile.ZipFile(f'../files/input/{file}', 'r') as z:
        with z.open(f'bank_marketing_{i}.csv') as f:
            if i == 0:
                df = pd.read_csv(f)
            else:
                df = pd.concat([df, pd.read_csv(f)], ignore_index=True)
df.tail(n=30)
```

```python
#
# Creación y limpiezade del dataframe df_client
#
df_client = df[['client_id', 'age', 'job', 'marital', 'education', 'credit_default', 'mortgage']]
df_client['job'] = df_client['job'].apply(lambda x: x.replace('.', '').replace('-', '_'))
df_client['education'] = df_client['education'].replace('.', '_').replace('unknown', pd.NA)
df_client['credit_default'] = df_client['credit_default'].replace('yes', 1).apply(lambda x: 0 if x != 1 else x)
df_client['mortgage'] = df_client['mortgage'].replace('yes', 1).apply(lambda x: 0 if x != 1 else x)
```

```python
#
# Creación del df_campaign y limpieza
#
df_campaign = df[['client_id', 'number_contacts', 'contact_duration', 'previous_campaign_contacts', 'previous_outcome', 'campaign_outcome']]
df_campaign['previous_outcome'] = df_campaign['previous_outcome'].replace('success', 1).apply(lambda x: 0 if x != 1 else x)
df_campaign['campaign_outcome'] = df_campaign['campaign_outcome'].replace('yes', 1).apply(lambda x: 0 if x != 1 else x)
    
df_campaign['last_contact_day'] = pd.to_datetime(df['day'].astype(str) + '-' + df['month'].astype(str) + '-2022')
df_campaign['last_contact_day'] = df_campaign['last_contact_day'].dt.strftime('%Y-%m-%d')
```
